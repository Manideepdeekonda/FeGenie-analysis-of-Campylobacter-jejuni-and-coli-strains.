```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
data = pd.read_csv('FeGenie_output/FeGenie-heatmap-data.csv')

# Display first few rows of the DataFrame
data.head()
```

```python
data.set_index('X', inplace=True)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data, annot=True, cmap='coolwarm', linewidths=.5)

# Customizing the plot
plt.title("FeGenie Heatmap", fontsize=20)
plt.xlabel("Gene sequences", fontsize=15)
plt.ylabel("Iron acquisition/regulation", fontsize=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
# Display the heatmap
plt.show()
```
